## Аппроксимирование функции

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

#Задаём исходную функцию
def f(x, a, b, c):
    f = c / (1 + ((x - a) / b) ** 2)
    return f

#Её коэффициенты
a = 0.5
b = 1
c = 2

xx = np.linspace(-3, 3, 50)
yy = f(xx, a, b, c)

#Генерация шума
np.random.seed(1200)
noise = np.random.normal(loc = 0 , scale = 0.03, size = xx.size) 
noisy_y = yy + 15 * noise

#Исходный график
plt.plot (xx, yy, '-')
plt.plot (xx, noisy_y, 'o')
plt.title ('Исходная функция с шумом')
plt.show()
```

```python
#Применение аппроксимации
popt, pcov= curve_fit(f, xx, noisy_y, [1, 1, 1])

#Конечный график с аппрокисмированной функцией
plt.plot (xx, yy, '--', label = 'Исходная функция')
plt.plot (xx, noisy_y, 'o', label = 'Шум') 
plt.plot (xx, f(xx, *popt), '-', label = 'Аппроксимированная функция')
plt.title ('Аппроксимированная функция')
plt.legend (loc = 'best' )
plt.show()
```
